# Start Stations (Weekday)

```python
def end_day(day):
    if day == "Monday" or day == "Tuesday" or day == "Wednesday" or day == "Thursday" or day == "Friday":
        return "Weekday"
    else:
        return "Weekend"
```

```python
# Name of the day
df["Start_End_Day"] = df.Start_NameOfDay.apply(end_day)
df["End_End_Day"] = df.Start_NameOfDay.apply(end_day)
```

```python
def part_of_day(hour):
    if (hour >= 7 and hour <= 9):
        return "Morning Rush Hour"
    elif (hour >= 16 and hour <= 18):
        return "Evening Rush Hour"
    else:
        return "Non Rush Hour"
```

```python
#Which part of the day?
df["Start_PartOfDay"] = df.Start_Hour.apply(part_of_day)
df["End_PartOfDay"] = df.End_Hour.apply(part_of_day)
```
